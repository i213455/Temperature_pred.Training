**Описание проекта**

Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. Ваша задача — построить модель, которая будет её предсказывать. Заказчик хочет использовать разработанную модель для имитации технологического процесса.
___

**Цель проекта**

Необходимо построить модель, имитирующую технологический процесс выплавки стали, и предсказывающую ее температуру. Целевой признак — последняя температура партии. Значение метрики MAE должно быть менее 6.8.
___

**Задачи проекта**

- Вывод
___

**Исходные данные**
Данные хранятся в Sqlite  — СУБД, в которой база данных представлена одним файлом `ds-plus-final.db`. Она состоит из нескольких таблиц:

- steel.data_arc — данные об электродах;
- steel.data_bulk — данные об объёме сыпучих материалов;
- steel.data_bulk_time — данные о времени подачи сыпучих материалов;
- steel.data_gas — данные о продувке сплава газом;
- steel.data_temp — данные об измерениях температуры;
- steel.data_wire — данные об объёме проволочных материалов;
- steel.data_wire_time — данные о времени подачи проволочных материалов.

1. Таблица steel.data_arc
    - key — номер партии;
    - BeginHeat — время начала нагрева;
    - EndHeat — время окончания нагрева;
    - ActivePower — значение активной мощности;
    - ReactivePower — значение реактивной мощности.
2. Таблица steel.data_bulk
    - key — номер партии;
    - Bulk1 … Bulk15 — объём подаваемого материала.
3. Таблица steel.data_bulk_time
    - key — номер партии;
    - Bulk1 … Bulk15 — время подачи материала.
4. Таблица steel.data_gas
    - key — номер партии;
    - gas — объём подаваемого газа.
5. Таблица steel.data_temp
    - key — номер партии;
    - MesaureTime — время замера;
    - Temperature — значение температуры.
6. Таблица steel.data_wire
    - key — номер партии;
    - Wire1 … Wire9 — объём подаваемых проволочных материалов.
7. Таблица steel.data_wire_time
    - key — номер партии;
    - Wire1 … Wire9 — время подачи проволочных материалов.

Во всех файлах столбец key содержит номер партии. В таблицах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

Перед нами стоит задача регрессии с метрикой mae не больше 6.8. Для такой точности лучше выбрать более сложные модели.

В качестве начальной возьмем линейную регрессию

Также используем случайный лес, градиентный бустинг и MLPRegressor. Для перебора гиперпараметров возьмем OptunaCV()

Так как все признаки числовые, то для предобработки используем только нормализацию

Перед обучением моделей выделим целевой признак

```python
y_train = X_train['temp_end']
X_train.drop(columns=['temp_end','Bulk10','Bulk3','PF'], inplace=True)

y_test = X_test['temp_end']
X_test.drop(columns=['temp_end','Bulk10','Bulk3','PF'], inplace=True)
```

```python
pipe_lr = Pipeline(
    steps = [("preprocessor", MinMaxScaler()),
             ("model_lr", LinearRegression())]
)

result_lr = -cross_val_score(pipe_lr, X_train, y_train, cv=6, scoring='neg_mean_absolute_error')

print('Среднее значение MAE трениовочной выборки при кросс-валидации:', result_lr.mean())
```

### Случайный лес <a id='22-link'></a>
Выберем модель случайного леса вместо дерева. При более долгом обучении она позволяет получить более точный результат

```python
pipe_forest = Pipeline([
    ('preprocessor', MinMaxScaler()),
    ('models', RandomForestRegressor(random_state=RANDOM_STATE))
])

parameters = {
        'models__min_samples_split': distributions.IntDistribution(60, 85),
        'models__min_samples_leaf': distributions.IntDistribution(2, 25),
        'models__max_depth': distributions.IntDistribution(10, 25)
}

oscv_forest = OptunaSearchCV(
    pipe_forest,
    parameters,
    scoring = 'neg_mean_absolute_error',
    cv=6,
    n_trials=10,
    random_state=RANDOM_STATE
)

oscv_forest.fit(X_train, y_train) 
forest_model = oscv_forest.best_estimator_


print('Лучшая модель и её параметры:\n\n', oscv_forest.best_params_)
print('Значение метрики лучшей модели:', -oscv_forest.best_score_)
```

### Модели градиентного бустинга <a id='23-link'></a>
Как наиболее точные и быстрые возьмем LGBMModel и CatBoostRegressor

```python
pipe_final = Pipeline([
    ('preprocessor', MinMaxScaler()),
    ('models', lgb.LGBMModel(objective="regression", random_state=RANDOM_STATE))
])

parameters = {
        'models__n_estimators': distributions.IntDistribution(25, 50)
}

oscv_LGBM = OptunaSearchCV(
    pipe_final,
    parameters,
    scoring = 'neg_mean_absolute_error',
    cv=6,
    n_trials=10,
    random_state=RANDOM_STATE
)

oscv_LGBM.fit(X_train, y_train) 
LGBM_model = oscv_LGBM.best_estimator_
```

```python
print('Лучшая модель LGBMModel и её параметры:\n\n', oscv_LGBM.best_params_)
print('Значение метрики лучшей модели LGBMModel:', -oscv_LGBM.best_score_)
```

```python
pipe_final = Pipeline([
    ('preprocessor', MinMaxScaler()),
    ('models', CatBoostRegressor(loss_function='MAE', random_state=RANDOM_STATE, learning_rate=0.15))
])

parameters = {'models__depth' : distributions.IntDistribution(2, 6),
              'models__iterations' : distributions.IntDistribution(65, 140)
              }

oscv_cat = OptunaSearchCV(
    pipe_final,
    parameters,
    scoring = 'neg_mean_absolute_error',
    cv=6,
    n_trials=10,
    random_state=RANDOM_STATE
)

oscv_cat.fit(X_train, y_train) 
cat_model = oscv_cat.best_estimator_
```

```python
print('Лучшая модель и её параметры CatBoostRegressor:\n\n', oscv_cat.best_params_)
print('Значение метрики лучшей модели CatBoostRegressor:', -oscv_cat.best_score_)
```
